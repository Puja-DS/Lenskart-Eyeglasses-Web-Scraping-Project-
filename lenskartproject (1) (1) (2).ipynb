```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://www.lenskart.com/eyeglasses.html")
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
Prod_Title = soup.select(".ProductTitle--xakon1.dZrMkC")
Prod_Title
```

```python
Product_Title=[]
for i in Prod_Title:
    Product_Title.append(i.get_text())
Product_Title
#len(Product_Title)
```

```python
Prod_S  = soup.select(".ProductSize--1gq7vf9.bTRlxU ")
Prod_S
```

```python
ProductS=[]
for i in Prod_S :
 ProductS.append(i.get_text())
ProductS
#len(Product_Strike_Price)
```

```python
product_elements = soup.select(".ProductSize--1gq7vf9.bTRlxU")
sizes = [element.get_text().split(' • ')[0].replace('Size: ', '') for element in product_elements]

# Display the extracted sizes
print("Sizes:", sizes)

#len(sizes)
```

```python
# Select the elements with the desired class
product_elements = soup.select(".ProductSize--1gq7vf9.bTRlxU")

# Extract the model/variant information
models = [element.get_text().split(' • ')[1] for element in product_elements]

# Display the model/variant information
print("Models/Variants:", models)
#len(models)
```

```python
ProductS=[]
for i in Prod_S:
 ProductS.append(i.get_text())
ProductS
#len(Prod_S)
```

```python
Prod_Price = soup.select(".SpecialPriceSpan--1mh26ry.ddtLHH")
Prod_Price
```

```python
Product_Pri=[]
for i in Prod_Price:
    Product_Pri.append(i.get_text())
Product_Pri
#len(Prod_Price)
```

```python
import re
Prices = [re.sub(r'[\u20B9]', '', price) for price in Product_Pri]
Prices 
```

```python
Prod_Rating = soup.select(".NumberedRatingSpan--kts3v6.gA-DJlF")
Prod_Rating
```

```python
Rating_list=[]
for i in Prod_Rating:
    Rating_list.append(i.get_text())
Rating_list
#len(Prod_Rating)
```

```python
offer = soup.select(".OfferContainer--gvtaaw.dQgfyT")
offer
```

```python
P_Offer=[]
for i in offer:
 P_Offer.append(i.get_text())
P_Offer
#len(offer)
```

```python
import pandas as pd
Eyeglasses = pd.DataFrame({
    "Product Name":Product_Title,
    "Size": sizes,
    "Model/Variant": models,
    "Price(Rs)":Prices,
    "Ratings(*)": Rating_list,
    "Special Offer": P_Offer
})
Eyeglasses
```

```python
Eyeglasses.to_csv("Eyeglasses_Web_Scraping.csv",index=False)
```
